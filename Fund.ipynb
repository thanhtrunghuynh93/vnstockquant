```python
import numpy as np
def check_performance(fund_name = "DCDS"):
    full_df = Fund().details.nav_report(fund_name)["nav_per_unit"]
    perf_df = full_df.pct_change().dropna()
    perf_df[perf_df > 0.07] = 0
    perf_df[perf_df < -0.07] = 0
    (perf_df + 1).cumprod().plot()
    CAGR, Sharpe, Sortino, PF, max_ddwn = analysis(perf_df)
    return perf_df
    
    
```

```python
import time
for i in range(20):
    fund_name = df["short_name"][i]
    print("Performance stats for:" + fund_name)
    check_performance(fund_name)
    print("=============================")
    time.sleep(10)
    
```
