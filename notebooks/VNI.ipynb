```python
# Annualized Sharpe Ratio
risk_free_rate_annual = 0.06   # e.g., 5%
risk_free_rate_daily = risk_free_rate_annual / 252

excess_daily_returns = daily_returns - risk_free_rate_daily
sharpe_ratio = np.mean(excess_daily_returns) / np.std(daily_returns, ddof=1)
annualized_sharpe = sharpe_ratio * np.sqrt(252)

print("Annualized Sharpe Ratio:", annualized_sharpe)
```
