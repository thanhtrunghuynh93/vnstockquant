```python
import pandas as pd
df = pd.read_csv("data/Vietnam_Manufacturing_PMI_2015_2025.csv")
df = df.dropna()
df = df.set_index("Date")
df.plot()
```

```python
df['PMI_3M_MA'] = df['PMI'].rolling(window=3).mean()
df['PMI_12M_MA'] = df['PMI'].rolling(window=12).mean()
df['PMI_Spread'] = df['PMI_3M_MA'] - df['PMI_12M_MA']

def determine_phase(row):
    if row['PMI'] > 50:
        return 'Expansion' if row['PMI_Spread'] > 0 else 'Downturn'
    else:
        return 'Recovery' if row['PMI_Spread'] > 0 else 'Contraction'

df['Economic_Cycle'] = df.apply(determine_phase, axis=1)
```
